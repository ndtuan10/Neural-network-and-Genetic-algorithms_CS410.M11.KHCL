```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [10, 20, 40, 80, 160]
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(16, 16))

#OneMax
# MRPS
ax = axs[0][0]

# 1X
y1X = [28.6, 81.2, 233.6, 608, 2201.6]
y1X_sde = [2.835, 18.766, 74.430, 77.066, 464.342] # standard deviation
# UX
yUX = [18.2, 31, 57.6, 98.4, 160.8]
yUX_sde = [3.376, 5.532, 5.987, 8.040, 16.179] # standard deviation

ax.errorbar(x, y1X, yerr=y1X_sde, uplims=True, lolims=True, label='1X')
ax.errorbar(x, yUX, yerr=yUX_sde, uplims=True, lolims=True, label='UX')
# ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('Problem size')
ax.set_ylabel('MRPS')
ax.set_title('MRPS of OneMax')
ax.legend()

# Evaluations
ax = axs[0][1]

# 1X
y1X = [225.06, 982.72, 4150.08, 16633.6, 89242.9]
y1X_sde = [21.376, 195.334, 1139.08, 1999.74, 16691.6] # standard deviation
# UX
yUX = [129.9, 318.06, 824.76, 2010, 4650.24]
yUX_sde = [27.460, 53.600, 81.586, 161.497, 453.557] # standard deviation

ax.errorbar(x, y1X, yerr=y1X_sde, uplims=True, lolims=True, label='1X')
ax.errorbar(x, yUX, yerr=yUX_sde, uplims=True, lolims=True, label='UX')
# ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('Problem size')
ax.set_ylabel('MRPS')
ax.set_title('Evaluation of OneMax')
ax.legend()

### Trap5

# MRPS
ax = axs[1][0]

# 1X
y1X = [121.2, 257.6, 921.6, 4019.2]
y1X_sde = [40.082, 53.761, 118.702, 801.816] # standard deviation
# UX
yUX = [361.6, 3532.8]
yUX_sde = [125.618, 665.6] # standard deviation

ax.errorbar([10,20,40,80], y1X, yerr=y1X_sde, uplims=True, lolims=True, label='1X')
ax.errorbar([10, 20], yUX, yerr=yUX_sde, uplims=True, lolims=True, label='UX')
# ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('Problem size')
ax.set_ylabel('MRPS')
ax.set_title('MRPS of Trap5')
ax.legend()

# Evaluations
ax = axs[1][1]

# 1X
y1X = [1069.6, 3368.8, 17661.4, 115023]
y1X_sde = [350.937, 708.442, 2323.43, 21493.9] # standard deviation
# UX
yUX = [4116.16, 82672.6]
yUX_sde = [1387.83, 15740.7] # standard deviation

ax.errorbar([10,20,40,80], y1X, yerr=y1X_sde, uplims=True, lolims=True, label='1X')
ax.errorbar([10, 20], yUX, yerr=yUX_sde, uplims=True, lolims=True, label='UX')
# ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('Problem size')
ax.set_ylabel('MRPS')
ax.set_title('Evaluation of Trap5')
ax.legend()

plt.show()

fig.savefig('visualize_plot.png', dpi=fig.dpi)
```
